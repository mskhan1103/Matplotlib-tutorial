```python
import matplotlib.pyplot as plt
```

```python
# Line plot
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Bar chart
labels = ['A', 'B', 'C']
values = [10, 20, 15]

plt.bar(labels, values)
plt.title("Bar Chart")
plt.show()
```

```python
# scatter plot
x = [1, 2, 3, 4]
y = [10, 5, 8, 12]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python
# Histogram
data = [1, 1, 2, 2, 2, 3, 4, 5, 5, 5, 5, 6]

plt.hist(data, bins=9)
plt.title("Histogram")
plt.show()
```

```python
# pie charts
labels = ['Apple', 'Banana', 'Cherry']
sizes = [40, 30, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
# Figures axes and subplots
fig, ax = plt.subplots()

ax.plot([1, 2, 3], [3, 2, 5])
ax.set_title("Title Here")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
plt.show()
```

```python
 # summary
Feature	Function/Command
Basic Plot	plt.plot(x, y)
Bar Chart	plt.bar()
Scatter Plot	plt.scatter()
Histogram	plt.hist()
Pie Chart	plt.pie()
Title & Labels	plt.title(), plt.xlabel(), etc.
Show Plot	plt.show()
Save Plot	plt.savefig()
Create Figure & Axes	fig, ax = plt.subplots()
```
